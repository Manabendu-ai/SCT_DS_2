```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("dataset/titanic/test.csv")
df
```

```python
df.isnull().sum()
```

```python
df["Age"] = df["Age"].fillna(df["Age"].median())
```

```python
df = df.drop(columns = ["Cabin"])
```

```python
df["Fare"] = df["Fare"].fillna(df["Fare"].median())
```

```python
df.duplicated().sum()
```

```python
df["PassengerId"] = df["PassengerId"].astype(str)
df["Ticket"] = df["Ticket"].astype(str)
```

```python
sns.histplot(df["Age"], kde = True, color = "orange")
plt.show()
```

```python
sns.boxplot(x = "Sex", y = "Fare", data=df, color = "green")
plt.show()
```

```python
numerical_df = df.select_dtypes(include=["number"])
sns.heatmap(numerical_df.corr(), annot=True, cmap = "coolwarm")
```

```python
df.groupby("Pclass")["Fare"].mean()
df.groupby("Sex")["Fare"].mean()
df.groupby("Embarked")["Fare"].mean()
```

```python
sns.boxplot(x = "Pclass", y = "Fare", data = df)
plt.title("Fare by Pclass")
plt.show()
```

```python
sns.violinplot(x = "Pclass", y = "Age", data = df, color = "purple")
plt.title("Age distribution across Pclass")
plt.show()
```

```python
sns.countplot(x = "Embarked", hue = "Pclass", data = df, palette = "dark:green")
plt.title("Passengers from each port by Pclass")
plt.show()
```

```python
sns.countplot(x = "Pclass", hue = "Sex", data = df, palette = "dark:orange")
plt.title("Gender distribution across Pclass")
plt.show()
```
